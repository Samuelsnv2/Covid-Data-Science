# Casos Confirmados de COVID-19

```python
import pandas as pd
from urllib.request import urlretrieve
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
urlretrieve(url, 'datasets/global_cases_covid19.csv')
```

```python
df_covid = pd.read_csv('datasets/global_cases_covid19.csv')
```

```python
df_covid
```

```python
type(df_covid)
```

```python
df_covid.head()
```

```python
df_covid.drop(['Lat','Long'], axis=1, inplace=True)
```

```python
df_covid.head()
```

```python
df_country = df_covid.groupby('Country/Region').sum()

df_country
```

```python
df_country.loc['Brazil']
```

```python
date = df_country.loc['Brazil'].index
cases = df_country.loc['Brazil'].values
```

```python
date
```

```python
cases
```

```python
plt.figure(figsize=(14,10))
plt.bar(date, cases)
```

```python
s_brazil = df_country.loc['Brazil']
```

```python
s_brazil
```

```python
s_brazil = s_brazil[s_brazil > 0]
```

```python
s_brazil.index
```

```python
def dates_values(s_country):
    tam = len(s_country)
    dates = s_country.index[tam-30:]
    values_t = s_country.values[tam-31:]
    values = []
    for i in range(len(values_t)):
        if i == 0:
            pass
        else:
            values.append(values_t[i] - values_t[i-1])
    return dates,values
```

```python
brazil_datas, valores_bra = dates_values(s_brazil)
plt.figure(figsize=(14,8))
plt.xticks(rotation=60)
plt.bar(brazil_datas, valores_bra)
plt.title('Casos confirmados por dia de COVID-19 no Brasil')
```

# Casos de Morte por COVID-19

```python
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
urlretrieve(url, 'datasets/global_cases_death_covid19.csv')
```

```python
df_covid = pd.read_csv('datasets/global_cases_death_covid19.csv')
```

```python
df_covid.drop(['Lat','Long'], axis=1, inplace=True)
```

```python
df_country = df_covid.groupby('Country/Region').sum()
```

```python
s_brazil = df_country.loc['Brazil']
```
