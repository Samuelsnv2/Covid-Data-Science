# Casos de Recuperados da Covid-19

```python
import pandas as pd
from urllib.request import urlretrieve
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
urlretrieve(url, 'datasets/global_cases_recovered_covid19.csv')
```

```python
df_covid = pd.read_csv('datasets/global_cases_recovered_covid19.csv')

df_covid
```

```python
df_covid.drop(['Lat','Long'], axis=1, inplace=True)
```

```python
df_country = df_covid.groupby('Country/Region').sum()

df_country
```

```python
s_brazil = df_country.loc['Brazil']

s_brazil
```

```python
s_brazil = s_brazil[s_brazil > 0]

s_brazil
```

```python
tam = len(s_brazil)
brazil_datas = s_brazil.index[tam-30:]
val_totais = s_brazil.values[tam-31:]
val_br = []
for i in range(len(val_totais)):
    if i == 0:
        pass
    else:
        val_br.append(val_totais[i] - val_totais[i-1])
val_br
#val_bra = []
#for i in val_br:
#    if i == 0:
#        pass
#    else:
#        val_bra.append(i)
#val_bra
```

```python
plt.figure(figsize=(14,8))
plt.xticks(rotation=60)
plt.bar(brazil_datas,val_br)
plt.title('Numero de Recuperados da COVID-19 nos ultimos 30 dias')
```
